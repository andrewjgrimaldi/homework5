```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_data = "data/mouse_drug_data.csv"
clinical_data = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
df_mouse = pd.read_csv(mouse_data)
df_clinical = pd.read_csv(clinical_data)

# Combine the data into a single dataset
df_mouse.head()
df_clinical.head()

df_merge = pd.merge(df_clinical, df_mouse, on="Mouse ID", how="left")
# Display the data table for preview
df_merge.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 

groupby0 = df_merge.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].mean()

# Convert to DataFrame

groupby1 = pd.DataFrame(groupby0.reset_index())

# Preview DataFrame
groupby1.head()

```

```python
groupby0.index
```

```python
groupby0.reset_index()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint (pandas.DataFrame.sem)

groupby2 = df_merge.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].sem()

# Convert to DataFrame . pd.DataFrame()

groupby3 = pd.DataFrame(groupby2.reset_index())

# Preview DataFrame

groupby3.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames - transforming the pivoting pivot_table().  https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.pivot_table.html

pt1 = groupby1.pivot_table(values = "Tumor Volume (mm3)", index = "Timepoint", columns = "Drug")

pt2 = groupby3.pivot_table(values = "Tumor Volume (mm3)", index = "Timepoint", columns = "Drug")

# Preview that Reformatting worked
pt1.head()
#pt2.head()
```

```python
plt.errorbar(pt1.index, groupby0["Capomulin"], yerr=pt2"Capomulin"], color="r", marker="o", markersize=5, linestyle="dashed", linewidth=0.50)
```

```python
# Generate the Plot (with Error Bars) - scatter plot.  df.scatter()

#pt1.plot()

plt.errorbar(pt1.index, pt1["Capomulin"], yerr=pt2["Capomulin"], color="r", marker="o", markersize=10, linestyle="solid", linewidth=0.90)
plt.errorbar(pt1.index, pt1["Infubinol"], yerr=pt2["Infubinol"], color="b", marker="^", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(pt1.index, pt1["Ketapril"], yerr=pt2["Ketapril"], color="g", marker="s", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(pt1.index, pt1["Placebo"], yerr=pt2["Placebo"], color="k", marker="d", markersize=5, linestyle="dashed", linewidth=0.50)

plt.title("Andrew's Chart")
plt.ylabel("Tumor Volume")
plt.xlabel("Time")
plt.legend()


# Save the Figure - ax.show()  .savefig(filelocation.csv).  adding a title and x and y axis.  ax.tile().  for the legend, ax.legend()

plt.savefig("homework5chart.jpg")
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 

mean_mat = df_merge.groupby(["Drug","Timepoint"])["Metastatic Sites"].mean()

# Convert to DataFrame

mean_mat = pd.DataFrame(mean_mat).reset_index()


# Preview DataFrame
mean_mat.head()
```

```python
df_merge.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 

sem_mean_mat = df_merge.groupby(["Drug","Timepoint"])["Metastatic Sites"].sem()

# Convert to DataFrame

sem_mean_mat = pd.DataFrame(sem_mean_mat).reset_index()

# Preview DataFrame
sem_mean_mat.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

pt3 = mean_mat.pivot_table(values = "Metastatic Sites", index = "Timepoint", columns = "Drug")
pt4 = sem_mean_mat.pivot_table(values = "Metastatic Sites", index = "Timepoint", columns = "Drug")


# Preview that Reformatting worked
pt3.head()
#pt4.head()
```

```python
pt3.index
```

```python
# Generate the Plot (with Error Bars)

plt.errorbar(pt3.index, pt3["Capomulin"], yerr=pt4["Capomulin"], color="k", marker="d", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(pt3.index, pt3["Ceftamin"], yerr=pt4["Ceftamin"], color="k", marker="d", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(pt3.index, pt3["Infubinol"], yerr=pt4["Infubinol"], color="k", marker="d", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(pt3.index, pt3["Ketapril"], yerr=pt4["Ketapril"], color="k", marker="d", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(pt3.index, pt3["Naftisol"], yerr=pt4["Naftisol"], color="k", marker="d", markersize=5, linestyle="dashed", linewidth=0.50)

plt.title("Andrew's 2nd Chart")
plt.ylabel("Metastatic thing")
plt.xlabel("Time")
plt.legend()


# Save the Figure
# Save the Figure - ax.show()  .savefig(filelocation.csv).  adding a title and x and y axis.  ax.tile().  for the legend, ax.legend()

plt.savefig("homework5chart.jpg")
plt.show()
# Show the Figure
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)

mice_groupby = df_merge.groupby(["Drug", "Timepoint"])["Mouse ID"].count()
# Convert to DataFrame

mice_df = pd.DataFrame(mice_groupby).reset_index()
# Preview DataFrame
mice_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

mice_df_pt = mice_df.pivot_table(values = "Mouse ID", index = "Timepoint", columns = "Drug")


# Preview the Data Frame
mice_df_pt.head()
```

```python
 # Generate the Plot (Accounting for percentages)
plt.plot(100 * mice_df_pt["Capomulin"] / 25, "ro", linestyle="dashed", markersize=5, linewidth=0.50)
plt.plot(100 * mice_df_pt["Infubinol"] / 25, "b^", linestyle="dashed", markersize=5, linewidth=0.50)
plt.plot(100 * mice_df_pt["Ketapril"] / 25, "gs", linestyle="dashed", markersize=5, linewidth=0.50)
plt.plot(100 * mice_df_pt["Placebo"] / 25 , "kd", linestyle="dashed", markersize=6, linewidth=0.50)
plt.plot(100 * mice_df_pt["Ceftamin"] / 25 , "yo", linestyle="dashed", markersize=6, linewidth=0.50)
plt.plot(100 * mice_df_pt["Naftisol"] / 25 , "cs", linestyle="dashed", markersize=6, linewidth=0.50)
plt.plot(100 * mice_df_pt["Propriva"] / 26 , "bd", linestyle="dashed", markersize=6, linewidth=0.50)
plt.plot(100 * mice_df_pt["Ramicane"] / 25 , "ko", linestyle="dashed", markersize=6, linewidth=0.50)
plt.plot(100 * mice_df_pt["Stelasyn"] / 25 , "bo", linestyle="dashed", markersize=6, linewidth=0.50)
plt.plot(100 * mice_df_pt["Zoniferol"] / 26 , "mo", linestyle="dashed", markersize=6, linewidth=0.50)


plt.title("Survival During Treatment")
plt.ylabel("Survival Rate (%)")
plt.xlabel("Time (Days)")
plt.grid(True)
plt.legend(loc="best", fontsize="small", fancybox=True)

# Save the Figure
plt.savefig("deadmouse.png")

# Show the Figure
plt.show()
```
